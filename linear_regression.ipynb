```python
# Splitting dataset into test/train
import pandas as pd
df_sal = pd.read_csv('/extracted_folder/Salary_Data.csv')
x = df_sal.iloc[:,:1]  # independent
y = df_sal.iloc[:, 1:]  # dependent
#print(x)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
print(x_train)
```

```python
# Regressor model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(x_test)     # predicted value of y_test
y_pred_train = regressor.predict(x_train)   # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(x_train, y_train, color = 'lightcoral')
plt.plot(x_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['x_train/Pred(y_test)', 'x_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Prediction on test set
plt.scatter(x_test, y_test, color = 'lightcoral')
plt.plot(x_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['x_train/Pred(y_test)', 'x_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(True)
plt.show()
```

```python
# Assuming the rest of your model training code is already in place

# Function to predict salary based on experience
def predict_salary(experience):
    # Convert the input experience into the format expected by the model (2D array)
    experience_array = [[experience]]
    # Use the trained model to predict the salary
    predicted_salary = regressor.predict(experience_array)
    return predicted_salary[0]

# Ask the user for their years of experience
user_experience = float(input("Enter your years of experience: "))

# Predict the salary based on the user input
predicted_salary = predict_salary(user_experience)

# Display the predicted salary
print(f"The predicted salary for {user_experience} years of experience is: {predicted_salary}")
```
