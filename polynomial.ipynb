```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Set the number of data points
m = 100  # We want to generate 100 data points

# Step 2: Generate random values for X
# np.random.rand(m, 1) generates 100 random values between 0 and 1
# We scale these values to be between -3 and 3 by multiplying by 6 and subtracting 3
X = 6 * np.random.rand(m, 1) - 3

# Step 3: Calculate the corresponding y values
# We use a quadratic equation: y = 0.5 * X^2 + X + 2
# Adding np.random.randn(m, 1) introduces some random noise
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)

# Step 4: Plot the data to visualize it
plt.scatter(X, y, color='blue', label='Generated Data')
plt.xlabel('X values')
plt.ylabel('y values')
plt.title('Generated Nonlinear Data')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Step 1: Generate nonlinear data
m = 100  # number of data points
X = 6 * np.random.rand(m, 1) - 3  # random X values in range [-3, 3)
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)  # quadratic equation with noise

# Step 2: Transform the data to include polynomial features (degree 2)
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Step 3: Fit the Linear Regression model to the polynomial features
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

# Step 4: Visualize the original data and the polynomial regression line
# Generate a range of values for X to plot the prediction curve
X_new = np.linspace(-3, 3, 100).reshape(100, 1)  # 100 points from -3 to 3
X_new_poly = poly_features.transform(X_new)  # transform these values to polynomial features
y_new = lin_reg.predict(X_new_poly)  # predict y values using the model

# Plotting
plt.scatter(X, y, color='blue', label='Original Data')  # scatter plot of original data
plt.plot(X_new, y_new, color='red', linewidth=2, label='Polynomial Regression Model')  # regression line
plt.xlabel('X values')
plt.ylabel('y values')
plt.title('Polynomial Regression Fit to Nonlinear Data')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Function to plot the polynomial regression
def plot_polynomial_regression(degree):
    # Step 1: Generate nonlinear data
    m = 100  # number of data points
    X = 6 * np.random.rand(m, 1) - 3  # random X values in range [-3, 3)
    y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)  # quadratic equation with noise

    # Step 2: Transform the data to include polynomial features
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly_features.fit_transform(X)

    # Step 3: Fit the Linear Regression model to the polynomial features
    lin_reg = LinearRegression()
    lin_reg.fit(X_poly, y)

    # Step 4: Visualize the original data and the polynomial regression line
    X_new = np.linspace(-3, 3, 100).reshape(100, 1)  # 100 points from -3 to 3
    X_new_poly = poly_features.transform(X_new)  # transform these values to polynomial features
    y_new = lin_reg.predict(X_new_poly)  # predict y values using the model

    # Plotting
    plt.scatter(X, y, color='blue', label='Original Data')  # scatter plot of original data
    plt.plot(X_new, y_new, color='red', linewidth=2, label=f'Polynomial Degree {degree}')  # regression line
    plt.xlabel('X values')
    plt.ylabel('y values')
    plt.title(f'Polynomial Regression Fit (Degree {degree})')
    plt.legend()
    plt.show()

# Example usage: visualize different polynomial degrees
for degree in [1, 2, 3, 5, 10]:
    plot_polynomial_regression(degree)
```
